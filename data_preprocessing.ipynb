```python
final_df = final_df[['id', 'title', 'genres', 'overview', 'keywords' , 'cast', 'crew']]
```

```python
import json
final_df['keywords'] = final_df['keywords'].apply(lambda x: json.loads(x))
final_df['cast'] = final_df['cast'].apply(lambda x: json.loads(x))
final_df['crew'] = final_df['crew'].apply(lambda x: json.loads(x))
final_df['genres'] = final_df['genres'].apply(lambda x: json.loads(x))
```

```python
final_df['cast'][0][0]['name']
```

```python
def find_director(row):
    li = []
    for i in row:
        if i['job'] == "Director":
            li.append(i['name'])
    return li
```

```python
def find_top_three_actors(row):
    li = []
    counter = 0
    for i in row:
        if counter < 3:
            li.append(i['name'])
            counter+=1
    return li
```

```python
def find_genre(row):
    li = []
    for i in row:
        li.append(i['name'])
    return li
```

```python
find_genre(final_df['genres'][0])
```

```python
find_top_three_actors(final_df['cast'][0])
```

```python
final_df['cast'] = final_df['cast'].apply(find_top_three_actors)
final_df['crew'] = final_df['crew'].apply(find_director) 
final_df['genres'] = final_df['genres'].apply(find_genre)
final_df['keywords'] = final_df['keywords'].apply(find_genre)
```

```python
final_df.head(1)
```

```python
def overview_list(row):
    return row.split()

overview_list(final_df['overview'][0])
```

```python
final_df['overview'] = final_df['overview'].apply(lambda x: str(x).split())
```

```python
final_df['tags'] = final_df['overview'] + final_df['genres'] + final_df['keywords'] + final_df['cast'] + final_df['crew']
```

```python

final_df.head(1)
```

```python
tags_df = final_df[['id', 'title', 'tags']]
```

```python
tags_df.head()
```

```python
tags_df['tags'] = tags_df['tags'].apply(lambda x: " ".join(x))
```

```python
tags_df.head()
```

```python
tags_df['tags'][0]
```

```python
tags_df['tags'] = tags_df['tags'].apply(lambda x: x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stemming(row):
    li = []
    for i in row.split():
        li.append(ps.stem(i))
    return " ".join(li)
tags_df['tags'] = tags_df['tags'].apply(stemming)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer( max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(tags_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similartiy = cosine_similarity(vectors)
```

```python
similartiy.shape
```

```python
def recommend(movie):
    
    movies_index = int(np.where(tags_df['title'] == movie)[0][0])
    distances = similartiy[movies_index]
    similar_movies = sorted(list(enumerate(distances)), key=lambda x:x[1], reverse=True)[1:6]
    for i in similar_movies:
        print(tags_df.iloc[i[0],1])
```
